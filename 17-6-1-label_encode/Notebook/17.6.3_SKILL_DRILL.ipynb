```python
# initial imports
import pandas as pd
from path import Path
```

```python
# Load data
file_path = Path("../Resources/loans_data.csv")
loans_df = pd.read_csv(file_path)
loans_df.head()
```

```python
# Binary encoding using Pandas (multiple columns)
loans_encoded_df = pd.get_dummies(loans_df, columns=["education", "gender"])
loans_encoded_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Creating an instance of label encoder
label_encoder = LabelEncoder()
loans_encoded_df["month_le"] = label_encoder.fit_transform(loans_df["month"])
loans_encoded_df.head()
```

```python
# Save to CSV
loans_encoded_df.to_csv(r'../Resources/loans_encoded.csv', index = False, header=True)
```
